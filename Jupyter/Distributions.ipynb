```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
small_uniform_sample = np.random.uniform(size =(100,1))
small_normal_sample = np.random.normal(size =(100,1))
small_binomial_sample = np.random.binomial(10,.5,100)


fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

ax1.hist(small_uniform_sample, bins = 50, color = 'k', alpha = .3)
ax1.set_title('Uniform Distribution')
ax2.hist(small_normal_sample, bins = 50, color = 'r', alpha = .3)
ax2.set_title('Normal Distribution')
ax3.hist(small_binomial_sample, bins = 50, color = 'y', alpha = .3)
ax3.set_title('Binomial Distribution')

plt.subplots_adjust(wspace=.3, hspace=.3)
```
